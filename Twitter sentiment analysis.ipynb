# Import Libraries

```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from textblob import TextBlob
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
text_df = df.drop(['id', 'user_name', 'user_location', 'user_description', 'user_created',
       'user_followers', 'user_friends', 'user_favourites', 'user_verified',
       'date', 'hashtags', 'source', 'retweets', 'favorites',
       'is_retweet'], axis=1)
text_df.head()
```

```python
print(text_df['text'].iloc[0],"\n")
print(text_df['text'].iloc[1],"\n")
print(text_df['text'].iloc[2],"\n")
print(text_df['text'].iloc[3],"\n")
print(text_df['text'].iloc[4],"\n")
```

```python
text_df.info()
```

```python
def data_processing(text):
    text = text.lower()
    text = re.sub(r"https\S+|www\S+https\S+", '',text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#','',text)
    text = re.sub(r'[^\w\s]','',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
nltk.download('punkt')
text_df.text = text_df['text'].apply(data_processing)
```

```python
text_df = text_df.drop_duplicates('text')
```

```python
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
text_df['text'] = text_df['text'].apply(lambda x: stemming(x))
```

```python
text_df.head()
```

```python
print(text_df['text'].iloc[0],"\n")
print(text_df['text'].iloc[1],"\n")
print(text_df['text'].iloc[2],"\n")
print(text_df['text'].iloc[3],"\n")
print(text_df['text'].iloc[4],"\n")
```

```python
text_df.info()
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df['polarity'] = text_df['text'].apply(polarity)
```

```python
text_df.head(10)
```

```python
def sentiment(label):
    if label <0:
        return "Negative"
    elif label ==0:
        return "Neutral"
    elif label>0:
        return "Positive"
```

```python
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
```

```python
text_df.head()
```

```python
vect = CountVectorizer(ngram_range=(1,2)).fit(text_df['text'])
```

```python

feature_names = vect.get_feature_names_out()
print("Number of features: {}\n".format(len(feature_names)))
print("First 20 features:\n {}".format(feature_names[:20]))
```

```python
X = text_df['text']
Y = text_df['sentiment']
X = vect.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print("Size of x_train:", (x_train.shape))
print("Size of y_train:", (y_train.shape))
print("Size of x_test:", (x_test.shape))
print("Size of y_test:", (y_test.shape))
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print(classification_report(y_test, logreg_pred))
```

```python
style.use('classic')
cm = confusion_matrix(y_test, logreg_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=logreg.classes_)
disp.plot()
```

```python
print("Best parameters:", grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
```

```python
logreg_acc = accuracy_score(y_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
import pandas as pd

data = pd.read_csv('vaccination_tweets.csv')

data['hashtags'] = data['hashtags'].apply(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
data['hashtags'] = data['hashtags'].str.replace('[','').str.replace(']','')
```

```python
hash_list = data.hashtags.tolist()
```

```python
hash_list
```

```python
import re
separated_words = []

for item in hash_list:
    if isinstance(item, str):  # Check if the item is a string
        words = re.findall(r'\w+', item)  # Use regular expression to find words
        separated_words.extend(words)  # Extend the result to the separated_words list

        
separated_words
```

```python
my_dict = {i:separated_words.count(i) for i in separated_words}
my_dict
```

```python
final = sorted(my_dict.items(), key=lambda x:x[1], reverse=True)
converted_dict = dict(final)
converted_dict
```

```python
top10 = {k: converted_dict[k] for k in list(converted_dict)[:10]}
top10
```

```python
keysList = list(top10.keys())
df_Trending = pd.DataFrame(keysList)
df_Trending.columns = ['Trends']
df_Trending
```

```python
import pickle

pickle.dump(df_Trending, open('Trending.sav','wb'))

trendsing = pickle.load(open('Trending.sav','rb'))
trendsing
```
